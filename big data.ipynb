# Using list comprehension, find names of employees who earn more than 7000

```python
columns=['name','department','age','salary','joining_date','created_date','last_updated_timestamp','rating','experience_years','bonus','project_count']
d=pd.read_csv("E:\DE\employee_data.csv",sep='~',usecols=columns)
#.set_index('id')
d
```

```python
#string to appropriate data types

dtypes1 = {
   
    'name':str,
    'age': int ,
    'salary': float,
    'department':str,
    'rating': float,
    'experience_years':int,
    'bonus': float,
    'project_count':int, 
    
}
```

```python
#string to appropriate data types
d=d.astype(dtypes1)
```

```python
# string to datetime 
d['joining_date']=pd.to_datetime(d['joining_date'])
d['created_date']=pd.to_datetime(d['created_date'])
d['last_updated_timestamp']=pd.to_datetime(d['last_updated_timestamp'])
```

```python
d.dtypes
```

```python
#Handling missing values, replace missing values in columns

d['age']=d['age'].fillna(0)
d['salary']=d['salary'].fillna(0)
d['joining_date']=d['joining_date'].fillna('1900-01-01')
d['bonus']=d['bonus'].fillna(0)
d['created_date']=d['created_date'].fillna('1900-01-01')
d['last_updated_timestamp']=d['last_updated_timestamp'].fillna('1900-01-01 00:00:00')
```
